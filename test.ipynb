```python
from requests_html import HTMLSession

s = HTMLSession()
url = "https://www.myntra.com/watches/maserati/maserati-men-black--silver-toned-chronograph-watch-r8873134003/13507600/buy"
r = s.get(url)
```

```python
import re

code = "<script>window.__myx = {...}</script>"

regex_pattern = r'window\.__myx\s*=\s*({.*?});'
match = re.search(regex_pattern, r.text)

if match:
    json_string = match.group(1)
    print(json_string)
```

```python
import json

json.loads(json_string)
```
